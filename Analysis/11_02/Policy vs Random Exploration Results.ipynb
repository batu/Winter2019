```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bbox_policy = np.load("bbox_policy.npy")
nn_policy = np.load("nn_policy.npy")

bbox_random = np.load("bbox_random.npy")
nn_random = np.load("nn_random.npy")
```

```python
bbox_policy_results = bbox_policy[:,-1]
nn_policy_results = nn_policy[:,-1]

bbox_random_results = bbox_random[:,-1]
nn_random_results = nn_random[:,-1]
```

```python
bbox_policy_mean = np.mean(bbox_policy_results)
bbox_policy_std = np.std(bbox_policy_results)

nn_policy_mean = np.mean(nn_policy_results)
nn_policy_std = np.std(nn_policy_results)

bbox_random_mean = np.mean(bbox_random_results)
bbox_random_std = np.std(bbox_random_results)

nn_random_mean = np.mean(nn_random_results)
nn_random_std = np.std(nn_random_results)

print(f"Policy BB mean: {bbox_policy_mean:0.1f} +- {bbox_policy_std:0.1f}")
print(f"Random BB mean: {bbox_random_mean:0.1f} +- {bbox_random_std:0.1f}")
print(f"\nPolicy NN mean: {nn_policy_mean:0.1f} +- {nn_policy_std:0.1f}")
print(f"Random NN mean: {nn_random_mean:0.1f} +- {nn_random_std:0.1f}")
```

```python
bbox_results = np.array([bbox_policy_mean, bbox_random_mean])
bbox_stds = np.array([bbox_policy_std, bbox_random_std])

nn_results = np.array([nn_policy_mean, nn_random_mean])
nn_stds = np.array([nn_policy_std, nn_random_std])
```

```python
x_pos = np.arange(len(bbox_results))
plt.bar(x_pos, bbox_results, yerr=bbox_stds, align='center', alpha=0.5)
plt.xticks(x_pos, ["Policy", "Random"])
plt.ylabel('Bounding Box Metric')
plt.title('Policy vs Random in RRT')
 
plt.show()
```

```python
x_pos = np.arange(len(nn_results))
plt.bar(x_pos, nn_results, yerr=nn_stds, align='center', alpha=0.5)
plt.xticks(x_pos, ["Policy", "Random"])
plt.ylabel('Nuclear Norm Metric Box Metric')
plt.title('Policy vs Random in RRT')
plt.show()
```
